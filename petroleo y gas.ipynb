```python
# Formatear eje Y para mostrar los valores completos
def millones(x, pos):
    return f'{int(x/1e6)}Mm'

plt.figure(figsize=(11, 6))
plt.plot(año, e_petroleo, label='Exportación de petróleo')
plt.gca().yaxis.set_major_formatter(FuncFormatter(millones))
plt.text(df.index[9], df.iloc[9,0], df.iloc[9,0], fontsize=11, color='red')
plt.text(df.index[2024-2005], df.iloc[2024-2005,0], df.iloc[2024-2005,0], fontsize=11, color='red')

plt.xlabel('Año')
plt.ylabel('Valor (en bariles)')
plt.title('Exportaciones Petróleo (bariles)')
plt.grid(True)
plt.xticks(año)
plt.legend()
```
